# demo01_random.ipynb  随机数

```python
import numpy as np
```

## 二项分布

```python
# 某人投篮命中率为0.3，投10次，进5个球的概率。
r = np.random.binomial(10, 0.5, 100000)
for i in range(11):
    print(i, ":",(r==i).sum() / 100000)
```

```python
# 某人打客服电话，客服接通率是0.6，一共打了3次，都没人接的概率。
r = np.random.binomial(3, 0.6, 100000)
for i in range(4):
    print(i, ":",(r==i).sum() / 100000)
```

## 超几何分布

```python
# 从6个好苹果、4个坏苹果中抽取3个苹果，返回好苹果的数量（执行10次）
r = np.random.hypergeometric(6, 4, 3, 100000)
for i in range(4):
    print(i, ":",(r==i).sum() / 100000)
```

## 饼状图

```python
import matplotlib.pyplot as plt

values = [15, 13.3, 8.5, 7.3, 4.62, 51.28]
spaces = [0.05, 0.01, 0.01, 0.01, 0.01, 0.01]
labels = ['Java', 'C', 'Python', 'C++', 'VB', 'Other']
colors = ['dodgerblue', 'orangered', 'limegreen', 'violet', 'gold','blue']

plt.axis('equal')
plt.pie(values, spaces, labels, colors, '%.1f%%', 
           shadow=True, startangle=90)
plt.legend()
```

## 统计数学成绩前100名 男女比例，绘制饼状图

```python
import pandas as pd
data=pd.read_csv('../data/学生考试表现数据/StudentsPerformance.csv')
mdata=data.sort_values(by='math score',ascending=False).head(100)

# 绘制饼状图
values = mdata['gender'].value_counts()
spaces = [0.05, 0.01]
labels = mdata['gender'].value_counts().index
colors = ['dodgerblue', 'orangered']
plt.axis('equal')
plt.pie(values, spaces, labels, colors, '%.1f%%', 
           shadow=True, startangle=90)
plt.legend()
```
