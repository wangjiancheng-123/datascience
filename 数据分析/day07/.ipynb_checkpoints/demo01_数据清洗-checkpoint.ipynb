## 检测与处理缺失值

```python
ratings = pd.read_json('../data/ratings.json')
# ratings[ratings.isnull()] = 3.5
# ratings
# ratings.isnull().sum().sum()  # 检测缺失值
ratings
```

```python
ratings.dropna()
ratings.dropna(axis=1, how='all')
ratings.dropna(axis=0, how='all', 
               subset=['Alex Roberts', 'Michael Henry'])
```

```python
# 填补（替换）法
ratings.fillna(3.5)
ratings.fillna(method='ffill')
ratings.fillna(method='ffill', axis=1)
```

```python
# 插值法
# 30 100注 40 120注 50 135注 60 155注 65 170注
x = np.array([30, 40, 50, 60, 65])
y = np.array([100, 120, 135, 155, 170])
# 生成插值器
import scipy.interpolate as si
si.interp1d(x, y, kind='linear')
```
