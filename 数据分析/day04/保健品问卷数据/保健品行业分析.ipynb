````python
# 切出前20列  做子集， 然后进行题目相关分析
sub_data = data.loc[:, :'q5_20']
corr = sub_data.corr()
corr[corr<0.3] = 0
corr[corr>=0.3] = 999
corr
````

结论：

2. 抗衰老
3. 有益身体健康
4. 外界看法

## 分析问卷调查的均值，了解当前各产品对产品功效的侧重点

````python
desc = sub_data.describe()
mean_score = desc.loc['mean'] 
mean_score.sort_values(ascending=False)
# 把题目（产品功效）追加到mean_score后
fields = pd.read_csv('保健品字段介绍.csv', encoding='gbk',
            header=None, names=['fid', 'effect', '_'],
            index_col='fid')
# 把effect与mean_score合并为一个DataFrame
fields = pd.DataFrame({ 'mean_score':mean_score, 
                        'effect':fields['effect']})
# 对dataFrame进行排序  按照mean_score降序排列
fields = fields.sort_values(by='mean_score', ascending=False)
fields
````

结论：
通过上表可知，多数产品在产品功效上更多关注改善用户的精神状态，其次为抗衰老，再次为改善身体健康，最后为外界影响。

## 从产品的角度进行分析

### 统计每一种保健品产品的频数，评估市场占有率

````python

# 整理产品名称
products = {1:'太太**',2:'乌鸡**',3:'排毒**',4:'阿*',5:'红桃*',6:'脑白*',7:'朵**',9:'其他'}
products = pd.Series(products)
result = pd.DataFrame({'products' : products, 
                       'counts'   : data['q8'].value_counts()})
result = result.sort_values(by='counts', ascending=False)
result
````

结论：
由上表可知，市场占有率最高的为“排毒”，第二梯队为“太太”、“红桃”、“朵”，针对当前这种情况，如果想要切入保健品市场，可以尽快抢占“乌鸡”、‘阿’、等产品的占有率，进而慢慢蚕食第二梯队的市场。所以后续分析可以关注一下这些品牌的用户群体。

### 研究每个产品功效的侧重点

````python
# 把每个产品当做一个小组，看一下每一组产品对20道题评分的最大值
result1 = pd.DataFrame([])
for k, v in products.items():
    mask = data['q8'] == k  # 当前组的掩码
    mean = sub_data[mask].mean()
    result1[k] = mean
result1
````

````python
result2 = result1.idxmax()
# 整理一个完整的 DataFrame
result2 = pd.DataFrame({'products' : products, 
                        'qid' : result2})

# series.apply(func)会遍历 series中的每个元素，交给func进行处理
# func函数会为每一个元素给出返回值， 作为apply方法的返回值返回
def func(item):
    return fields.effect.loc[item]

result2['effect'] = result2['qid'].apply(func)
result2
````

结论：
根据每个产品的功效可得：若希望切入保健品市场，在功效方面侧重于用户的精神状态的改善，而针对‘乌鸡’、‘阿’等产品可以从广告投放，舆论扩散，产品口号等等角度抢占他们的市场。

## 从用户群体角度分析

### 研究每个产品的用户群体，已婚、未婚频数统计

````python
result3 = pd.DataFrame([])
for k, v in products.items():
    mask = data['q8'] == k
    r = data[mask]['w1'].value_counts()
    result3[k] = r
result3 = result3.T
result3.columns = ['married', 'not-married']
result3['products'] = products
result3
````

结论：
由上表可知，“太太”、‘乌鸡’、‘阿’更多服务已婚女士，而未婚女士更多会选择“排毒”、‘红桃’。由此可知，已婚女士会更多的注重保健品的护肤、抗衰老等功效，未婚女士更多关注服务之后的精神状态。针对“乌鸡”、‘阿’等角度分析，可以更多的关注已婚女士对保健品的需求，设计相应的产品形象。
